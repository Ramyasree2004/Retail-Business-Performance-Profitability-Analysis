```python
# Retail Business Performance & Profitability Analysis - Blinkit Data

## Step 1: Data Loading and Cleaning

import pandas as pd

# Load data
df = pd.read_excel("BlinkIT Grocery Data.xlsx")

# Fill missing values
df['Item Weight'].fillna(df['Item Weight'].mean(), inplace=True)

# Standardize categorical values
df['Item Fat Content'] = df['Item Fat Content'].replace({
    'low fat': 'Low Fat', 'LF': 'Low Fat', 'reg': 'Regular'
})


## Step 2: Exploratory Data Analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot of Sales by Item Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='Item Type', y='Sales', data=df)
plt.xticks(rotation=90)
plt.title("Sales by Item Type")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
